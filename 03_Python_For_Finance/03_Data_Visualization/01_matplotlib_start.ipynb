```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(1000)
y = np.random.standard_normal(20)
```

```python
x = range(len(y))
plt.plot(x,y)
```

```python
plt.plot(y)
plt.plot(y.cumsum())
plt.grid(True)
plt.axis('tight')
```

```python
plt.figure(figsize=(7,4)) # size of the figure (width, height)
plt.plot(y.cumsum(), 'g', lw=1.5)
plt.plot(y.cumsum(), 'ro')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

```python
np.random.seed(2000)
y = np.random.standard_normal((20,2)).cumsum(axis=0)
y[:,0] = y[:,0] * 100
y
```

```python
fig, ax1 = plt.subplots() # plot first dataset using left axis
plt.plot(y[:,0], 'b', lw=1.5, label='1st')
plt.plot(y[:,0], 'ro')
plt.grid(True)
plt.legend(loc=8)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value 1st')
ax2 = ax1.twinx() # plot second dataset using right axis
plt.plot(y[:,1], 'g', lw=1.5, label='2nd')
plt.plot(y[:,1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```
